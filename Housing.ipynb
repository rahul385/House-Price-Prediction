# Predicting California Housing Prices

This project aims to predict the final sale price of houses based on different explanatory variables describing aspects. Predicting house prices is useful to identify fruitful investments, or to determine whether the price advertised for a house is over or underestimated, before making a buying judgment.

The dataset is based on data from the 1990 California census.

```python
# to handle dataset
import pandas as pd
import numpy as np

# to profile the data
from pandas_profiling import ProfileReport

# to visualize data
import matplotlib.pyplot as plt
import seaborn as sns

# to split dataset
from sklearn.model_selection import train_test_split

# feature scaling
from sklearn.preprocessing import StandardScaler

# to train model
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor

# to evaluate model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

### Separate train and test set

```python
# Separate data into train and test set
X_train, X_test, y_train, y_test = train_test_split(df, df.median_house_value, test_size=0.3,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

### Imputation: Replace missing data

```python
# Check Missing data
X_train.info()
```

```python
# Imputation: add variable indicating missingness + median imputation
# Feature importance insights: Variable indicating missingness is least important feature. So don't create this feature
for df in [X_train, X_test]:
    for var in ['total_bedrooms']:
        #df[var+'_NA'] = np.where(df[var].isnull(), 1, 0)
        df[var].fillna(X_train[var].median(), inplace=True)
```

```python
# Validate imputation
X_train.info()
X_test.info()
```

### Feature Extraction

```python
# Feature Extraction. Check new features in training set
X_train['rooms_per_bedroom']=X_train['total_rooms']/X_train['total_bedrooms']
X_train['population_per_household']=X_train['population']/X_train['households']
X_train["rooms_per_household"] = X_train["total_rooms"]/X_train["households"]
a=X_train.corrwith(X_train['median_house_value'])
a.sort_values(ascending=False)
```

```python
# population_per_household is not correlated so drop it
X_train=X_train.drop(['population_per_household'], axis=1)
```

```python
# Feature Extraction. Add new features in test set
X_test['rooms_per_bedroom']=X_test['total_rooms']/X_test['total_bedrooms']
X_test["rooms_per_household"] = X_test["total_rooms"]/X_test["households"]
```

```python
# Validate new features
X_train.head()
X_test.head()
```

### Engineering Rare labels

```python
# Categorical features
X_train.ocean_proximity.value_counts()
```

```python
# Replace island with most frequent feature
X_train=X_train.replace(to_replace =["ISLAND"],value ="<1H OCEAN") 
X_test=X_test.replace(to_replace =["ISLAND"],value ="<1H OCEAN") 
```

### Feature Encoding

```python
# Feature Encoding: One Hot Encoding for ocean_proximity

X_train=pd.concat([X_train[['longitude','latitude','housing_median_age','total_rooms','total_bedrooms',
'population','households','median_income','median_house_value','rooms_per_bedroom',
'rooms_per_household']],
pd.get_dummies(X_train['ocean_proximity'], drop_first=True)],axis=1)
                    
X_test=pd.concat([X_test[['longitude','latitude','housing_median_age','total_rooms','total_bedrooms',
'population','households','median_income','median_house_value','rooms_per_bedroom'
,'rooms_per_household']],
pd.get_dummies(X_test['ocean_proximity'], drop_first=True)],axis=1)
```

```python
# Validate OHE
X_train.head()
X_test.head()
```

### Outliers: Detection & Handling

```python
# Outliers
# let's make boxplots to visualise outliers in the continuous variables 
# and histograms to get an idea of the distribution
continuous=('longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income',
'median_house_value','rooms_per_bedroom','rooms_per_household')

for var in continuous:
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    fig = X_train.boxplot(column=var)
    fig.set_title('')
    fig.set_ylabel(var)
    
    plt.subplot(1, 2, 2)
    fig = X_train[var].hist(bins=20)
    fig.set_ylabel('Number of houses')
    fig.set_xlabel(var)

    plt.show()
```

```python
def tree_binariser(var):
    score_ls = [] # here I will store the mse

    for tree_depth in [1,2,3,4]:
        # call the model
        tree_model = DecisionTreeRegressor(max_depth=tree_depth)

        # train the model using 3 fold cross validation
        scores = cross_val_score(tree_model, X_train[var].to_frame(), y_train, cv=3, scoring='neg_mean_squared_error')
        score_ls.append(np.mean(scores))

    # find depth with smallest mse
    depth = [1,2,3,4][np.argmax(score_ls)]
    print(score_ls, np.argmax(score_ls), depth)

    # transform the variable using the tree
    tree_model = DecisionTreeRegressor(max_depth=depth)
    tree_model.fit(X_train[var].to_frame(), X_train.median_house_value)
    X_train[var] = tree_model.predict(X_train[var].to_frame())
    X_test[var] = tree_model.predict(X_test[var].to_frame())
```

```python
continuous=('total_rooms','total_bedrooms','population','households','median_income'
            ,'rooms_per_bedroom','rooms_per_household')
for var in continuous:
    tree_binariser(var)
```

```python
# Validate discretization
X_train.head()
X_test.head()
```

```python
training_vars = [var for var in X_train.columns if var not in ['median_house_value']]
training_vars
```

### Feature scaling

```python
# fit scaler
scaler = StandardScaler() # create an instance
scaler.fit(X_train[training_vars]) #  fit  the scaler to the train set for later use
```

### Model Training and Evaluation

##### Linear Regression

```python
# Train Model Linear Regression
lin_model=LinearRegression()
lin_model.fit(scaler.transform(X_train[training_vars]),y_train)

# RMSE
Y_train_pred = lin_model.predict(scaler.transform(X_train[training_vars]))
rms = np.sqrt(mean_squared_error(y_train, Y_train_pred))
print('Linear Regression RMS:',rms)

# Cross Validation Score
scores = cross_val_score(lin_model, X_train[training_vars], y_train,
                             scoring="neg_mean_squared_error", cv=5)
rmse_scores = np.sqrt(-scores)
print("Cross Validation Mean:", rmse_scores.mean())
print("Standard deviation:", rmse_scores.std())
```

##### Decision Tree

```python
# Train Model Decision Tree
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X_train[training_vars], y_train)

# RMSE
Y_train_pred = tree_model.predict(X_train[training_vars])
rms = np.sqrt(mean_squared_error(y_train, Y_train_pred))
print('Decision Tree RMSE:',rms)

# Cross Validation Score
scores = cross_val_score(tree_model, X_train[training_vars], y_train,
                             scoring="neg_mean_squared_error", cv=5)
rmse_scores = np.sqrt(-scores)
print("Cross Validation Mean:", rmse_scores.mean())
print("Standard deviation:", rmse_scores.std())
```

##### Random Forest

```python
# Train Model Random Forest
rf_model = RandomForestRegressor()
rf_model.fit(X_train[training_vars], y_train)

# RMSE
Y_train_pred = rf_model.predict(X_train[training_vars])
rms = np.sqrt(mean_squared_error(y_train, Y_train_pred))
print('Random Forest RMSE:',rms)

# Cross Validation Score
scores = cross_val_score(rf_model, X_train[training_vars], y_train,
                             scoring="neg_mean_squared_error", cv=5)
rmse_scores = np.sqrt(-scores)
print("Cross Validation Mean:", rmse_scores.mean())
print("Standard deviation:", rmse_scores.std())
```

##### XGBoost

```python
# Train Model XGBoost
xgb_model = xgb.XGBRegressor()
xgb_model.fit(X_train[training_vars], y_train)

# RMSE
Y_train_pred = xgb_model.predict(X_train[training_vars])
rms = np.sqrt(mean_squared_error(y_train, Y_train_pred))
print('XGB RMSE:',rms)

# Cross Validation Score
scores = cross_val_score(xgb_model, X_train[training_vars], y_train,
                             scoring="neg_mean_squared_error", cv=5)
rmse_scores = np.sqrt(-scores)
print("Cross Validation Mean:", rmse_scores.mean())
print("Standard deviation:", rmse_scores.std())
```

### Feature Importance

```python
importance = pd.Series(rf_model.feature_importances_)
importance.index = training_vars
importance.sort_values(inplace=True, ascending=False)
importance.plot.bar(figsize=(18,6))
```

```python
importance = pd.Series(xgb_model.feature_importances_)
importance.index = training_vars
importance.sort_values(inplace=True, ascending=False)
importance.plot.bar(figsize=(18,6))
```

### Optimal Model

```python
# Final Model --> XGBoost
Y_test_pred = xgb_model.predict(X_test[training_vars])
rms = np.sqrt(mean_squared_error(y_test, Y_test_pred))
print('XGBoost RMSE:', rms)
```
